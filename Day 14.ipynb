# Prepare data

## Question: Show me total revenue by category

```python
df = spark.sql("""
    SELECT 
        category_id, 
        SUM(price) AS total_revenue
    FROM ecommerce.default.ecommerce_full_data
    GROUP BY category_id
    ORDER BY total_revenue DESC
""")
df.limit(5).display()
```

## Which products have the highest conversion rate?

```python
df_conversion = spark.sql("""
    SELECT 
        product_id, 
        Conversion_rate
    FROM ecommerce.default.gold_ecommerce_full_data
    ORDER BY Conversion_rate DESC
    LIMIT 5
""")
display(df_conversion)
```

## What's the trend of daily purchases over time

```python
df_daily_purchases = spark.sql("""
    SELECT 
        event_date, 
        COUNT(*) AS num_purchases
    FROM ecommerce.default.ecommerce_full_data
    WHERE event_type = 'purchase'
    GROUP BY event_date
    ORDER BY event_date
""")
display(df_daily_purchases)
```

## Find customers who viewed but never purchased

```python
df_never_purchased = spark.sql("""
    SELECT 
        user_id
    FROM ecommerce.default.ecommerce_full_data
    WHERE event_type = 'view'
      AND user_id NOT IN (
          SELECT DISTINCT user_id
          FROM ecommerce.default.ecommerce_full_data
          WHERE event_type = 'purchase'
      )
    GROUP BY user_id
""")
display(df_never_purchased)
```
