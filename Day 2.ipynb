```python
# Reading a ecommerce data to a dataframe
df = spark.read.csv('/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv',header=True,inferSchema=True,sep=',')
df.limit(5).display()
```

```python
# Reading first 5 rows
df.limit(5).show()
```

```python
# Slicing first 3 columns
df.select(df.columns[:3]).limit(5).display()
```

```python
# Slicing first 3 columns
df.select(df.columns[:3]).show(5)
```

```python
df.select(df.columns[:3]).take(2)
```

Filter

```python
df.filter(df["price"] > 1000).count()
```

```python
df.groupBy("brand").agg({"price":"avg"}).show(5)
```

```python
top_brands = df.filter("brand is not null").groupBy("brand").count().orderBy("count",ascending=False).limit(5)
top_brands.display()
```

```python
df.createOrReplaceTempView("ecommerce")
```

```python
# Querying the temporary view using PySpark
spark.sql("""
select user_id,category_code,brand,price from ecommerce 
where price > 100 
order by price desc
limit 5
""").show()
```
