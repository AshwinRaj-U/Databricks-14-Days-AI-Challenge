```python
# Creating a Df from CSV file
df = spark.read.csv('/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv',header=True,inferSchema=True)
df.limit(5).display()
```

```python
# Converting existing df to delta
df.write.format('delta').mode('overwrite').save('/Volumes/workspace/ecommerce/ecommerce_data/delta/2019_Nov')
```

```python
# Create managed table
# This gets stored in the default warehouse
df.write.format("delta").saveAsTable("2019_Nov")
```

```python
try:
    wrong_schema = spark.createDataFrame([("a","b","c")], ["x","y","z"])
    wrong_schema.write.format("delta").mode("append").save("/Volumes/workspace/ecommerce/ecommerce_data/delta/2019_Nov")
except Exception as e:
    print(f"Schema enforcement: {e}")
```
